```python
df['Age_mean_inputed'] = df['Age'].fillna(df['Age'].mean())
df['Age_median_inputed'] = df['Age'].fillna(df['Age'].median())

df['Cabin_mode_inputed'] = df['Cabin'].fillna(df['Cabin'].mode()[0])

df[['Age','Age_mean_inputed','Age_median_inputed','Cabin','Cabin_mode_inputed']]
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.boxplot(data = df, x ='Pclass', y = 'Age', palette = 'pastel')

plt.title('Boxplot of Age by Passenger Class', fontsize = 16)
plt.xlabel('Passenger Class', fontsize = 12)
plt.ylabel('Age', fontsize = 12)

plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.boxplot(data = df, x ='Embarked', y = 'Fare', palette = 'pastel')

plt.title('Boxplot of Fare by Embarked', fontsize = 16)
plt.xlabel('Embarked', fontsize = 12)
plt.ylabel('Fare', fontsize = 12)

plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

encoded_data = df.copy()

label_encoder = LabelEncoder()
encoded_data['Sex_Label'] = label_encoder.fit_transform(encoded_data['Sex'])

one_hot_encoded = pd.get_dummies(encoded_data['Embarked'], prefix='Embarked')
encoded_data = pd.concat([encoded_data, one_hot_encoded], axis=1)

pclass_mapping = {1: 'First', 2: 'Second', 3: 'Third'}
encoded_data['Pclass_Category'] = encoded_data['Pclass'].map(pclass_mapping)

print(encoded_data[['Sex','Sex_Label','Embarked','Embarked_Q','Embarked_S','Embarked_C','Pclass']])
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

numerical_columns = ['Age', 'Fare', 'Pclass']

df.dropna(subset=numerical_columns, inplace=True)

min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()
robust_scaler = RobustScaler()

df['Age_MinMax'] = min_max_scaler.fit_transform(df[['Age']])
df['Fare_MinMax'] = min_max_scaler.fit_transform(df[['Fare']])

df['Age_Standard'] = standard_scaler.fit_transform(df[['Age']])
df['Fare_Standard'] = standard_scaler.fit_transform(df[['Fare']])

df['Age_Robust'] = robust_scaler.fit_transform(df[['Age']])
df['Fare_Robust'] = robust_scaler.fit_transform(df[['Fare']])

print(df[['Age','Age_MinMax','Age_Standard','Age_Robust']])
```
