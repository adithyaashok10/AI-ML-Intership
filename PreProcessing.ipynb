```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Age_mean_inputed'] = df['Age'].fillna(df['Age'].mean())
df['Age_median_inputed'] = df['Age'].fillna(df['Age'].median())

df[['Age','Age_mean_inputed','Age_median_inputed']]
```

```python
sns.set(style='whitegrid')

plt.figure(figsize=(20,6))
sns.boxplot(data = df, x = 'Age', y = 'Performance_Rating', palette='pastel')

plt.title('Performance vs Age')
plt.xlabel('Age')
plt.ylabel('Performance')

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

encoded_data = df.copy()

label_encoder = LabelEncoder()
encoded_data['Gender_Label'] = label_encoder.fit_transform(encoded_data['Gender'])

one_hot_encoded = pd.get_dummies(encoded_data['Department'], prefix='Department')
encoded_data = pd.concat([encoded_data, one_hot_encoded], axis=1)

perfomance_mapping={1:'First',2:'Second',3:'Third',4:'four',5:'five'}
encoded_data['Performance_Rating_Label'] = encoded_data['Performance_Rating'].map(perfomance_mapping)

print(encoded_data[['Department_Finance','Department_HR','Department_IT','Department_Marketing','Department_Sales']])
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

numerical_coloumns = ['Age','Experience_Years','Monthly_Salary','Performance_Rating']

min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

df['Age_Standard'] = standard_scaler.fit_transform(df[['Age']])
df['Experience_Years_standard'] = standard_scaler.fit_transform(df[['Experience_Years']])
df['Monthly_Salary_Standard'] = standard_scaler.fit_transform(df[['Monthly_Salary']])
df['Performance_Rating_Standard'] = standard_scaler.fit_transform(df[['Performance_Rating']])

df['Age_MinMax'] = min_max_scaler.fit_transform(df[['Age']])
df['Experience_Years_MinMax'] = min_max_scaler.fit_transform(df[['Experience_Years']])
df['Monthly_Salary_MinMax'] = min_max_scaler.fit_transform(df[['Monthly_Salary']])
df['Performance_Rating_MinMax'] = min_max_scaler.fit_transform(df[['Performance_Rating']])

print(df[['Age_Standard','Experience_Years_standard','Monthly_Salary_Standard','Performance_Rating_Standard','Age_MinMax','Experience_Years_MinMax','Monthly_Salary_MinMax','Performance_Rating_Standard']])
```
